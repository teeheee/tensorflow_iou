```python

import matplotlib.pyplot as plt
import matplotlib
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

def plot(intersec):
    patches = []

    fig, ax = plt.subplots()

    patches.append(Polygon(intersec, fill=True))
    patches.append(Polygon(rect1, fill=False))
    patches.append(Polygon(rect2, fill=False))


    p = PatchCollection(patches, cmap=matplotlib.cm.jet, alpha=0.4)

    colors = 100*np.random.rand(len(patches))
    p.set_array(np.array(colors))

    ax.add_collection(p)

    plt.xlim([0,8])
    plt.ylim([-5,5])
    plt.show()
    
print("cpp")
plot(intersection_cpp)
print("python")
plot(intersection)
```
